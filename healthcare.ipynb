```python
def categorize_age(a):
    if a > 60:
        return 'Senior Citizens >60'
    elif a > 23:
        return 'Citizens >23'
    elif a >= 18:
        return 'Adults >=18'
    elif a >= 10:
        return 'Minors <18'
    elif a < 10:
        return 'Kids <10'
    else:
        return 'NA'

df["Category"] = df["Age"].apply(categorize_age)
```
