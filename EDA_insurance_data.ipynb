```python
# Convert categorical columns to numerical values
df_encoded = df.copy()

# Map gender, diabetic, and smoker to numeric values
df_encoded['gender'] = df_encoded['gender'].map({'male': 1, 'female': 0})
df_encoded['diabetic'] = df_encoded['diabetic'].map({'Yes': 1, 'No': 0})
df_encoded['smoker'] = df_encoded['smoker'].map({'Yes': 1, 'No': 0})

# Drop non-numeric columns that are not needed for the correlation
df_encoded = df_encoded.drop(['index', 'PatientID', 'region'], axis=1)

# Calculate the correlation matrix
corr_matrix = df_encoded.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

plt.title('Correlation Heatmap')
plt.show()
```
